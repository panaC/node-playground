```typescript
import * as fetch from 'node-fetch'
```

```typescript
let db;

fetch('https://github.com/jshttp/mime-db/raw/master/db.json')
    .then(res => res.json())
    .then(json => db = json);
```

```typescript
let mimeTypes: any;
```

```typescript
mimeTypes = Object.entries(db)
    .filter(([key, value]) => (value as any)?.extensions?.length)
    .reduce((pv, [key, value]) => [...pv, [key, (value as any).extensions]], [])
    .reduce((pv, [key, value]) => [...pv, (value as string[]).map((v) => [key, v])], [])
    .reduce((pv, cv) => [...pv, ...cv] , [])
    .reduce((pv, [key, value]) => ({...pv, [key]: value}), {})
```
